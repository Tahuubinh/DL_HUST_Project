```python
for i in data_of_a_city.columns.tolist():
    if (i == 'Date'):
        continue
    data_of_a_city[i] = (data_of_a_city[i].ffill() + data_of_a_city[i].bfill())/2
    data_of_a_city[i] = data_of_a_city[i].bfill().ffill()
```

```python
data_of_a_city
```

```python
data_of_a_city = data_of_a_city.interpolate(method='nearest')
data_of_a_city.to_csv('./Delhi_air_quality_nearest_interpolate.csv')
```
