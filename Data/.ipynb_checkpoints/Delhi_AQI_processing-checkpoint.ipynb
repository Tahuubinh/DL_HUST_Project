```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

data = pd.read_csv('./Delhi_air_quality_remove_index_col.csv')
matplotlib.style.use('ggplot')
figure(figsize=(6, 20))
```

```python
data.head(10)
```

```python
data.tail(10)
```

```python
def getMaxNaNIndex(data, column):
    print(data[column].index[data[column].apply(np.isnan)].max())
for i in data.columns.tolist():
    if i == 'Datetime' or i == 'City':
        continue
    if i == 'AQI_Bucket':
        break
    print(f'{i}: ', end = '')
    getMaxNaNIndex(data, i)
```

```python
df = data.set_index('Datetime')
df
```

```python
df["AQI"].plot(rot=90, figsize=(15,5))
```

```python
for i, col in enumerate(df.columns):
    plt.figure(i)
    if col != 'AQI':
        ax = df[col].plot(rot=90, figsize=(15,5))
    else:
        ax = df[col].plot(rot=90, figsize=(15,5), color = 'blue')
    ax.set_ylabel(col)
```
